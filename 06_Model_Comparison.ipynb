# Credit Card Fraud Detection - Model Comparison

Compare saved models using precision, recall, F1-score, and ROC curves.


```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score, roc_curve, auc
```

```python
with open('processed_data.pkl', 'rb') as fh:
    data = pickle.load(fh)
models = {}
for filename in ['logistic_regression.pkl', 'decision_tree.pkl', 'knn_model.pkl']:
    with open(f'../models/{filename}', 'rb') as fh:
        models[filename.replace('.pkl', '')] = pickle.load(fh)
```

```python
rows = []
plt.figure(figsize=(8, 6))
for name, model in models.items():
    y_pred = model.predict(data['X_test'])
    y_score = model.predict_proba(data['X_test'])[:, 1]
    fpr, tpr, _ = roc_curve(data['y_test'], y_score)
    plt.plot(fpr, tpr, label=f'{name} (AUC={auc(fpr, tpr):.3f})')
    rows.append({'model': name, 'precision': precision_score(data['y_test'], y_pred), 'recall': recall_score(data['y_test'], y_pred), 'f1': f1_score(data['y_test'], y_pred)})
plt.plot([0, 1], [0, 1], 'k--')
plt.legend()
plt.title('ROC comparison')
plt.show()
pd.DataFrame(rows).sort_values('f1', ascending=False)
```
