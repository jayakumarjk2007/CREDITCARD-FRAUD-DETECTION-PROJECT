# Credit Card Fraud Detection - Data Exploration

Exploratory data analysis for imbalance, transaction amount, time, and feature correlations.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
```

```python
df = pd.read_csv('../creditcard.csv')
print(f'Dataset shape: {df.shape}')
df.head()
```

```python
print('Missing values:', df.isna().sum().sum())
print('Duplicate rows:', df.duplicated().sum())
df.info()
```

```python
class_counts = df['Class'].value_counts().sort_index()
print(class_counts)
print(f'Fraud rate: {class_counts.loc[1] / len(df) * 100:.4f}%')
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
class_counts.plot(kind='bar', ax=axes[0], color=['#2ca25f', '#de2d26'])
axes[0].set_title('Class distribution')
axes[0].set_yscale('log')
axes[0].set_xticklabels(['Legitimate', 'Fraud'], rotation=0)
axes[1].pie(class_counts, labels=['Legitimate', 'Fraud'], autopct='%1.4f%%', colors=['#a1d99b', '#fc9272'])
axes[1].set_title('Class share')
plt.tight_layout()
```

```python
df.groupby('Class')['Amount'].describe()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', center=0)
plt.title('Feature correlation matrix')
plt.tight_layout()
```
