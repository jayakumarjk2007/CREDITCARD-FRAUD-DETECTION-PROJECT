# Credit Card Fraud Detection - Decision Tree

Train a tree-based classifier and inspect feature importance.


```python
import pickle
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
with open('processed_data.pkl', 'rb') as fh:
    data = pickle.load(fh)
model = DecisionTreeClassifier(max_depth=10, random_state=42)
model.fit(data['X_train_smote'], data['y_train_smote'])
```

```python
y_pred = model.predict(data['X_test'])
y_score = model.predict_proba(data['X_test'])[:, 1]
print(classification_report(data['y_test'], y_pred))
print(confusion_matrix(data['y_test'], y_pred))
print('ROC-AUC:', roc_auc_score(data['y_test'], y_score))
```

```python
importance = pd.DataFrame({'feature': data['X_train_smote'].columns, 'importance': model.feature_importances_})
importance.sort_values('importance', ascending=False).head(10)
```

```python
with open('../models/decision_tree.pkl', 'wb') as fh:
    pickle.dump(model, fh)
print('Saved ../models/decision_tree.pkl')
```
