# Credit Card Fraud Detection - Model Evaluation

Evaluate the selected model with confusion matrix, precision-recall curve, and financial impact.


```python
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, precision_recall_curve
from utils import calculate_financial_impact
```

```python
with open('../models/decision_tree.pkl', 'rb') as fh:
    model = pickle.load(fh)
with open('processed_data.pkl', 'rb') as fh:
    data = pickle.load(fh)
```

```python
y_pred = model.predict(data['X_test'])
y_score = model.predict_proba(data['X_test'])[:, 1]
cm = confusion_matrix(data['y_test'], y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(data['y_test'], y_score)
plt.plot(recall, precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.grid(alpha=0.3)
plt.show()
```

```python
calculate_financial_impact(data['y_test'], y_pred)
```
