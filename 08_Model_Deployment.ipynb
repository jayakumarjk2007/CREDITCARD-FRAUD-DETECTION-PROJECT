# Credit Card Fraud Detection - Model Deployment

Load the trained model and expose a reusable prediction function.


```python
import pickle
import pandas as pd
```

```python
with open('../models/decision_tree.pkl', 'rb') as fh:
    model = pickle.load(fh)
```

```python
def predict_fraud(transaction_frame):
    prediction = model.predict(transaction_frame)
    probability = model.predict_proba(transaction_frame)[:, 1]
    return {'prediction': int(prediction[0]), 'fraud_probability': float(probability[0])}
```

```python
with open('processed_data.pkl', 'rb') as fh:
    data = pickle.load(fh)
sample = data['X_test'].iloc[[0]]
predict_fraud(sample)
```
