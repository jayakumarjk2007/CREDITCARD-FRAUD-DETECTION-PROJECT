# Credit Card Fraud Detection - Logistic Regression

Train and evaluate a linear baseline model.


```python
import pickle
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, RocCurveDisplay
import matplotlib.pyplot as plt
```

```python
with open('processed_data.pkl', 'rb') as fh:
    data = pickle.load(fh)
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(data['X_train_smote'], data['y_train_smote'])
```

```python
y_pred = model.predict(data['X_test'])
y_score = model.predict_proba(data['X_test'])[:, 1]
print(classification_report(data['y_test'], y_pred))
print(confusion_matrix(data['y_test'], y_pred))
print('ROC-AUC:', roc_auc_score(data['y_test'], y_score))
```

```python
RocCurveDisplay.from_predictions(data['y_test'], y_score)
plt.title('Logistic Regression ROC Curve')
plt.show()
```

```python
with open('../models/logistic_regression.pkl', 'wb') as fh:
    pickle.dump(model, fh)
print('Saved ../models/logistic_regression.pkl')
```
